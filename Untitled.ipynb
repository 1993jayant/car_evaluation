### Plotting the distribution of each feature

```python
# function for printing disribution/count of a feature/column and plotting the same
def disb_plot(col):
    '''Print the count and frequency of each category in a feature/label. It plots the frequency distribution also using bar plots.'''
    
    disb = data.groupby(col).size() # data['col'].value_counts() can also be used here
    print(disb, '\n\n************************************')
    
    # percentage of data points of each category
    print(f'\nFrequency of different values is:\n{(disb/disb.sum())*100}')
    
    # plotting the distribution of class labels
    plt.bar(x= disb.index, height = disb, label = 'Count of each label')
    plt.xlabel(f'{col}')
    plt.ylabel('Count of labels')
    plt.title('Distribution of labels')
    plt.grid()
    plt.legend()
    plt.show()
```

### Feature wise class distribution

```python
# category wise distribution of class labels
def eda(col, cat, feat):
    # creating the filter
    filt = (data[col] == cat)
    val_cnt = data.loc[filt, feat].value_counts()
    print(f'Absolute values for "{feat}" column in "{col}\'s" "{cat}" category is:\n', val_cnt, '\n***********************************\n')
    print(f'Percentage wise distribution of "{cat}" category in "{col}" column for each "{feat}" is:\n{(val_cnt/val_cnt.sum())*100}\n***************************************')
    
    # plotting the distribution using bar plots
    plt.bar(x = val_cnt.index, height = val_cnt, label = f'Count of "{feat}"')
    plt.title(f'Distribution of each "{feat}" for "{col}\'s" "{cat}" category')
    plt.xlabel(f'{feat}')
    plt.ylabel('Count')
    plt.legend()
    plt.grid()
    plt.show()
```

### Data Preprocessing

```python
# creating a new dataframe with preprocessed features. replacing strings with numerical values

ppro_data = pd.DataFrame()
ppro_data['buying_price'] = data['buying_price'].replace(['vhigh', 'high', 'med', 'low'], [0, 1, 2, 3])
ppro_data['maintenance'] = data['maintenance'].replace(['vhigh', 'high', 'med', 'low'], [0, 1, 2, 3])
ppro_data['doors'] = data['doors'].replace(['2', '3', '4', '5more'], [0, 1, 2, 3])
ppro_data['persons'] = data['persons'].replace(['2', '4', 'more'], [0, 1, 2])
ppro_data['lug_boot'] = data['lug_boot'].replace(['small', 'med', 'big'], [0, 1, 2])
ppro_data['safety'] = data['safety'].replace(['low', 'med', 'high'], [0, 1, 2])
ppro_data['class'] = data['class'].replace(['unacc', 'acc', 'good', 'vgood'], [0, 1, 2, 3])
```

```python
# separating the class label
y = ppro_data['class']

# dropping the class label from dataframe
ppro_data.drop(['class'], axis = 1, inplace = True)
```

```python
# balancing the class weight

class_names = data['class'].unique()

# computing class weights to make each class balanced
class_weight = compute_class_weight('balanced', class_names, data['class'])

# making dictionary of class_names and their corresponding weights
zipp = dict(zip(class_names, class_weight))

for name in class_names:
    filt = data['class'] == name
    ppro_data.loc[filt] *= zipp[name]
```

##### Data Splitting

```python
# splitting the data in 70:30 ratio
xtrain, xtest, ytrain, ytest = train_test_split(ppro_data, y, test_size = 0.3, stratify = y)
```

```python
# encoding the class label into OHE format
lb = LabelBinarizer()
y_ohe = lb.fit_transform(y)
```

```python
# splitting the data in 70:30 ratio in one vs rest
xtrain_ovr, xtest_ovr, ytrain_ovr, ytest_ovr = train_test_split(ppro_data, y_ohe, test_size = 0.3, stratify = y)
```

### Data Modelling

#### KNN

```python
# Finding the optimal hyperparameter using GridSearchCV
knn = KNeighborsClassifier()
params = dict(n_neighbors = [1, 5, 10, 20, 40, 80, 160])
grid = GridSearchCV(estimator = knn, cv = 5, return_train_score = True, param_grid = params, scoring = 'roc_auc')
grid.fit(xtrain, ytrain)
```

```python
# plotting the AUC score
plt.plot(params['n_neighbors'] , grid.cv_results_['mean_train_score'], label = 'Train_auc_score')
plt.plot(params['n_neighbors'], grid.cv_results_['mean_test_score'], label = 'Test_auc_score')
plt.scatter(params['n_neighbors'] , grid.cv_results_['mean_train_score'], label = 'Train_auc_score')
plt.scatter(params['n_neighbors'], grid.cv_results_['mean_test_score'], label = 'Test_auc_score')
plt.xlabel('no of neighbors')
plt.ylabel('Auc_score')
plt.title('AUC plot of KNN')
plt.grid()
plt.legend()
plt.show()
```

```python
# training the model with optimal hyperparameter
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(xtrain, ytrain)
```

```python
# accuracy on test data
knn.score(xtest,ytest)
```

```python
# number of classes identified by the knn model
knn.classes_
```

* We can see that KNN has trained four models using **one vs rest** technique automatically.

#### Logistic Regression

```python
# Finding the optimal hyperparameter for logistic regression
c = [0.1, 1, 10, 20, 40, 80, 160]

score_tr = []
score_te = []
for val in c:
    lr = LogisticRegression(C = val,multi_class = 'multinomial')
    lr.fit(xtrain, ytrain)
    score_tr.append(lr.score(xtrain, ytrain))
    score_te.append(lr.score(xtest, ytest))
```

```python
# plotting the accuracy on train and test data
plt.plot(c, score_tr, label = 'train_score')
plt.plot(c, score_te, label = 'test_score')
plt.scatter(c, score_tr, label = 'train_score')
plt.scatter(c, score_te, label = 'test_score')
plt.xlabel('Hyperparameter')
plt.ylabel('Accuracy')
plt.title('Train and Test accuracy')
plt.grid()
plt.legend()
plt.show()
```

```python
# Training the final model with optimal hyperparameter
lr = LogisticRegression(C = 10, multi_class = 'ovr')
lr.fit(xtrain, ytrain)
print(lr.coef_)
print('\n Accuracy on test data is: ',lr.score(xtest, ytest))
```

* Interpretability is not as easy for multiclass problem as for binary classification. As we can see , for class '2', i.e., for class 'good', the polarity or sign of coefficient terms is reversed. This is because the model has considered it as class 0(in one vs rest setting).

#### Naive Bayes

```python
# Finding he optimal hyperparameter for naive bayes
alpha = [0.0001, 0.001, 0.01, 0.1, 1, 2]

train = []
test = []
for i in alpha:
    nb = MultinomialNB(alpha = i)
    nb.fit(xtrain, ytrain)
    train.append(nb.score(xtrain, ytrain))
    test.append(nb.score(xtest, ytest))

# Plotting the accuracy on train and test data
plt.plot(alpha, train, label = 'Train_accuracy')
plt.plot(alpha, test, label = 'Test_accuracy')
plt.scatter(alpha, train)
plt.scatter(alpha, test)
plt.legend()
plt.grid()
plt.xlabel('Alpha')
plt.ylabel('Accuracy')
plt.title('Naive Bayes accuracy plot')
plt.show()
```

```python
# Fitting the model with optimal hyperparameter
nb = MultinomialNB(alpha = 0.0001)
nb.fit(xtrain, ytrain)

# printing the accuracy score on test data
print('Accuracy on test data is : ',nb.score(xtest, ytest))
```

* Naive Bayes is giving very less accuracy, so we can discard this model.

#### SVM

```python
# Finding best hyperparameter using gridsearchcv
svm = SVC()
c = [1, 10, 20, 40, 80, 160]
gama = [0.1, 0.2, 0.4, 0.8, 1.6]
params = dict(C = c, gamma = gama)
grid= GridSearchCV(estimator = svm, return_train_score = True, cv = 5, scoring = 'accuracy', param_grid = params) # roc_auc is only available for binary classification task
grid.fit(xtrain, ytrain)
```

```python
# Plotting the heatmap for each combination of hyperparameter
sns.heatmap(grid.cv_results_['mean_test_score'].reshape(6,5), annot = True, fmt = 'f', xticklabels = gama, yticklabels = c, linewidths = 0.1)
plt.title('Accuracy for each hyperparameter combination')
plt.xlabel('Hyperparameter "gamma"')
plt.ylabel('Hyperparameter "C"');
```

```python
# best hyperparameter combination
grid.best_params_
```

```python
# Best score we are getting is
grid.best_score_
```

```python
# fitting the model with optimal hyperparameters

svm = SVC(C = 10, gamma = 0.1)
svm.fit(xtrain, ytrain)

# printing the accuracy score on train and test data
print('Training accuracy is: ', svm.score(xtrain, ytrain))
print('Testing accuracy is: ', svm.score(xtest, ytest))
```

* As we can see that we are getting very high accuracy score by using SVM with RBF kernel.

#### Decision Tree

```python
# finding the optimal hyerparameter for decision tree

dt = DecisionTreeClassifier(min_samples_leaf = 20)
mx_dep = [2, 3, 4, 5, 10]
min_sampl = [5, 10, 20, 40, 80, 160]
params = dict(max_depth = mx_dep, min_samples_split = min_sampl)
grid = GridSearchCV(estimator = dt, cv = 5, return_train_score = True, scoring = 'roc_auc', param_grid = params)
grid.fit(xtrain_ovr, ytrain_ovr) # this algo is accepting the class labels to be in OHE format
```

```python
# best hyperparameters
grid.best_params_
```

```python
# heatmap of the roc_score for each combination of hyperparameter
sns.heatmap(grid.cv_results_['mean_test_score'].reshape(5, 6), annot = True, fmt = 'f', xticklabels = min_sampl, yticklabels = mx_dep, linewidths = 0.1)
plt.xlabel('min_samples_split')
plt.ylabel('max_depth')
plt.title('AUC score for decision tree');
```

```python
# fitting the model with optimal hyperparameter
dt = DecisionTreeClassifier(max_depth = 3, min_samples_split = 5, min_samples_leaf = 20)
dt.fit(xtrain_ovr, ytrain_ovr)

print('Accuracy score for train data is: ',dt.score(xtrain_ovr, ytrain_ovr))
print('Accuracy score for test data is: ',dt.score(xtest_ovr, ytest_ovr))
```

* We can see that we have got perfect score for accuracy using Decision Trees.

#### Random Forest

```python
# Finding the best hyperparameter using gridsearchcv
rf = RandomForestClassifier()
n_estim = [1, 5, 10, 20]
params = dict(n_estimators = n_estim)
grid = GridSearchCV(estimator = rf, scoring = 'roc_auc', param_grid = params, return_train_score = True)
grid.fit(xtrain_ovr, ytrain_ovr)
```

```python
# optimal hyperparameter
grid.best_params_
```

```python
# AUC score for test data
grid.score(xtest_ovr, ytest_ovr)
```

```python
plt.plot(n_estim, grid.cv_results_['mean_train_score'], label = 'train AUC score')
plt.plot(n_estim, grid.cv_results_['mean_test_score'], label = 'Test AUC score')
plt.scatter(n_estim, grid.cv_results_['mean_train_score'])
plt.scatter(n_estim, grid.cv_results_['mean_test_score'])
plt.title('AUC score for Random Forest')
plt.xlabel('No of estimators')
plt.ylabel('AUC score')
plt.legend()
plt.grid()
plt.show()
```

* Random forest is also giving perfect AUC score.

#### XGBoost

```python
# Finding the optimal hyperparameter for xgboost model
xgb = XGBClassifier()
n_estim = [1, 3, 5, 10, 20, 40]
learn_rate = [0.1, 0.3]
params = dict(n_estimators = n_estim, learning_rate = learn_rate)
grid = GridSearchCV(estimator = xgb, return_train_score = True, param_grid = params)
grid.fit(xtrain, ytrain)
```

```python
# best combination of hyperparameters
grid.best_params_
```

```python
# heatmap for each combination of hyperparameters
sns.heatmap(grid.cv_results_['mean_test_score'].reshape(6, 2), annot = True, fmt = 'f', xticklabels = learn_rate, yticklabels = n_estim, linewidths = 0.1)
plt.title('Accuracy score for XGBoost model')
plt.ylabel('number of estimators')
plt.xlabel('learning_rate')
plt.show()
```

* We are getting perfect accuracy score for all combination of hyperparameters.
